```python
df = df.drop_duplicates(keep="first")
```

```python
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'])
```

```python
df['VehicleIntroDate'] = pd.to_datetime(df['VehicleIntroDate'])
```

```python
df['make'] = df['make'].str.strip()
```

```python
columns_onehot = ["IsVATRegistered", "Citizenship", "LegalType", "Title", "Bank", "AccountType", "MaritalStatus", "Gender", "Province", "MainCrestaZone", "SubCrestaZone", "VehicleType", "make", "Model", "bodytype", "AlarmImmobiliser", "TrackingDevice", "CapitalOutstanding", "NewVehicle", "TermFrequency", "ExcessSelected", "CoverCategory", "CoverType", "CoverGroup", "Section", "Product"]

df_OHE = oneHotEncoder( df, columns_onehot)
```

```python
columns_label = ['Language','Country','ItemType' , 'StatutoryClass', 'StatutoryRiskType' ]
df_encode= labelEncoder(df_OHE, columns_label)
```

```python
columns_scaler = ["PostalCode", "mmcode", "RegistrationYear", "Cylinders", "cubiccapacity", "kilowatts", "NumberOfDoors", "SumInsured", "CalculatedPremiumPerTerm", "TotalClaims"]
df_standard = Scaler(df_encode, columns_scaler)
df_standard
```

```python
X = df_standard.drop('TotalPremium', axis=1)
y = df_standard['TotalPremium']
X_train, X_test, y_train, y_test = split_data(X, y)
```

```python
date_col = ["VehicleIntroDate", "TransactionMonth"]

for col in date_col:
    X_train['year'] = X_train[col].dt.year
    X_train['month'] = X_train[col].dt.month
    X_train['day'] = X_train[col].dt.day

    # Drop the original DateTime column if no longer needed
    X_train.drop(columns=[col], inplace=True)
```

```python
X_train.dtypes.unique()
```

```python
X_train = X_train.astype('float64')
```

```python
lr_model, dt_model, rfr_model, xgb_model = train_models(X_train, y_train)
```

```python
# Evaluate models
models = ['Linear Regression', 'Decision Tree', 'Random Forest', 'XGBoost']
mae_scores, mse_scores, r2_scores = [], [], []
for model in [lr_model, dt_model, rfr_model, xgb_model]:
    mae, mse, r2, _ = evaluate_model(model, X_test, y_test)
    mae_scores.append(mae)
    mse_scores.append(mse)
    r2_scores.append(r2)

# Plot metrics
plot_metrics(models, mae_scores, mse_scores, r2_scores)
```
